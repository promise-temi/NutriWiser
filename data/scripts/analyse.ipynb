```python
import pandas as pd
```

```python
df1 = pd.read_csv('../samples/sample1.csv', on_bad_lines="skip")
df2 = pd.read_csv('../samples/sample2.csv', on_bad_lines="skip")
df3 = pd.read_csv('../samples/sample3.csv', on_bad_lines="skip")
df4 = pd.read_csv('../samples/sample4.csv', on_bad_lines="skip")
df5 = pd.read_csv('../samples/sample5.csv', on_bad_lines="skip")
df6 = pd.read_csv('../samples/sample6.csv', on_bad_lines="skip")
df7 = pd.read_csv('../samples/sample7.csv', on_bad_lines="skip")

df = pd.concat([df1, df2, df3, df4, df5, df6, df7], ignore_index=True)
```

```python
import missingno as msno
```

```python
msno.matrix(df)
```

```python
seuil = 95

colonnes_completement_vides = df.isna().sum() / len(df) * 100 > seuil
colonnes_completement_vides_liste = colonnes_completement_vides[colonnes_completement_vides].index.tolist()
print(len(colonnes_completement_vides_liste))
colonnes_completement_vides_liste
```

```python
df_sans_colonnes_vides = df.drop(columns=colonnes_completement_vides_liste)
```

```python
msno.matrix(df_sans_colonnes_vides)
```
